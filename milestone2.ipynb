```python
# Read in neccessary libraries
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df_food = pd.read_csv("USDA_data.csv")
df_food.head()
```

Next, we need to tidy the data imported from the USDA Food Access Research Atlas. Below is a tidied dataframe with columns named more appropriately and unneccessary columns of data removed. This new dataframe displays data surrounding county population, total number of housing units, number of houses without vehicle access, median household income, poverty rate, and access to food of houses at 1 mile, 10 miles, and 20 miles from a grocery store.

```python
# Tidy data by creating new dataframe containing only the relevant parameters
df_food = df_food[["State","County","POP2010",
      "OHU2010",
      "PovertyRate",
      "MedianFamilyIncome", 
      "CensusTract",
      "lapop1share","lapop10share","lapop20share","TractHUNV", "LATracts1","LATracts10","LATracts20"]]

df_food.rename(columns={"POP2010": "Population", "OHU2010": "TotalHouseUnits","TractHUNV":"HouseUnitsNoVehicle"}, inplace=True)

display(df_food.head())
df_food.dtypes

#Change data types to proper format
df_food.astype({'LATracts1': 'object','LATracts10': 'object','LATracts20': 'object'}).dtypes
```

Below I have created a pivot table indexed on state and county to organize the above dataframe into a more manageable format.

```python
# Create a pivot table to organize data by State and County
df_food_pivot = df_food.pivot_table(values=['Population','TotalHouseUnits','PovertyRate','MedianFamilyIncome','lapop1share',
                                            'lapop10share','lapop20share','HouseUnitsNoVehicle','LATracts1','LATracts10','LATracts20'],
                                   index=['State','County'],
                                   aggfunc={'Population':np.sum,'TotalHouseUnits':np.sum,'HouseUnitsNoVehicle':np.sum, 'PovertyRate':np.mean,'MedianFamilyIncome':np.mean,'lapop1share':np.mean,
                                            'lapop10share':np.mean,'lapop20share':np.mean,'LATracts1':np.mean,'LATracts10':np.mean,'LATracts20':np.mean})
df_food_pivot.round(2)
```

I decided to only use a fraction of the parameters available in this dataset to start and make the data more manageable. I find these parameters including, census tract number, state, county, population count, count of housing units, poverty rate, median family income, low access population at 1 mile, 10 miles, and 20 miles, and vehicle access, to be the most useful at this point. I formatted the dataframe to be grouped by state and county.

#### County Health Data

```python
# Read in data from University of Wisconsin Population Health Institute 
df_health = pd.read_csv("County_Health.csv", header=[1])

# Calculate the percentage of premature deaths from the overall population of each county
df_health['% Premature Deaths'] = df_health.Deaths/df_health['Aggregate Population']

# Select columns to keep in dataframe
df_health = df_health[['State','County','Aggregate Population','% Premature Deaths','% Fair/Poor',
                       'Physically Unhealthy Days','Mentally Unhealthy Days','% Smokers','% Obese','% Uninsured',
                       '% unemployed','% Children in Poverty','GINI','% Healthy Food']]
df_health.rename(columns={'% Fair/Poor':'% Fair/Poor Health',
                          '% Healthy Food':'% of Zip Codes with Healthy Food Access',
                          'GINI':'Gini coefficient of household income inequality'}, inplace= True)
display(df_health)
df_health.dtypes
```

I decided to only use a fraction of the parameters available in this dataset to start and make the data more manageable. I find these parameters including, state, county, aggregate population, percentage of premature deaths, percentage of fair/poor health days in a month, count of physically and mentally unhealthy days a month, percent of smokers, percent obese, percent uninsured adults, percent unemployed adults, percent of children in poverty, and percent of zip codes in a given county with access to healthy food, to be the most useful at this point. I formatted the dataframe to be grouped by state and county. Then I created a pivot table to make the data more organized.

```python
# Create pivot table to index by State and County
df_health_pivot = df_health.pivot_table(values=['Aggregate Population',
                                       '% Premature Deaths','% Fair/Poor Health','Physically Unhealthy Days',
                                       'Mentally Unhealthy Days','% Smokers','% Obese','% Uninsured',
                                               '% unemployed','% Children in Poverty','Gini coefficient of household income inequality', '% of Zip Codes with Healthy Food Access'],
                                       index=['State','County'])
display(df_health_pivot)
```

#### Combining USDA Food Access Data and County Health Data

```python
# Using .merge() function to combine the two dataframes into one 
df = df_food.merge(df_health, left_on=['State','County'], right_on=['State','County'], how='inner')
display(df)
# Create pivot table of new dataframe indexed by 'State' and 'County'
df_pivot = df.pivot_table(values=['Population','TotalHouseUnits','PovertyRate','MedianFamilyIncome','lapop1share',
                                'lapop10share','lapop20share','HouseUnitsNoVehicle','LATracts1','LATracts10','LATracts20',
                                '% Premature Deaths','% Fair/Poor Health','Physically Unhealthy Days','Mentally Unhealthy Days',
                                '% Smokers','% Obese','% Uninsured','% unemployed','% Children in Poverty',
                                'Gini coefficient of household income inequality', '% of Zip Codes with Healthy Food Access'],
                         index=['State','County'])

display(df_pivot.round(2))
```

## Exploratory Data Analysis

To get a general understanding of our data and identify which parameters have the largest effect on a communities overall health, I decided to create a correlation matrix of all of the parameters when relating to obesity rates and a combined overall unhealthy days parameter that I created from the data presenting the average number of unhealthy days an individual reported per month in each county. The overall unhealthy days parameter is calculated as the sum of the average physically unhealthy days per month and the average mentally unhealthy days per month of an individual, divided by 30 to get the percentage of the month an individual is mentally and/or physicallly unhealthy. 

First, lets look at which parameters have the greatest correlation when using obesity rate as an indicator of community overall health. 

```python
# Find correlation between each variable and obesity rates
corr_df = df_pivot.corr()
display(corr_df)

# Display the correlation values in order in a bar graph
obesity_corr_df = corr_df[['% Obese']]["% Obese"]
obesity_corr_df = obesity_corr_df.drop('% Obese')
obesity_corr_df = obesity_corr_df.sort_values()

# Choose the greatest correlation parameters from all parameters considered 
greatest_obesity_corr_df = obesity_corr_df[(obesity_corr_df >= .3) | (obesity_corr_df <= -.3)]
greatest_obesity_corr_df = greatest_obesity_corr_df.sort_values()
```

```python
# Plot all correlation values and the greatest correlation value parameters for Obesity rate
fig, ax = plt.subplots(1,2, figsize=(15,5))

obesity_corr_df.plot.bar(title = 'Correlation Values for Obesity Rate',ax=ax[0])
greatest_obesity_corr_df.plot.bar(title = 'Greatest Value Correlation Parameters for Obesity Rate',ax=ax[1], color='red')
```

The two bar graphs above suggest that median family income, the percent of the population that is unemployed, the number of physically unhealthy days, the percent of smokers in a population, the poverty rate, the percent of children in poverty, the percent of premature deaths in a population, and the percentage of individuals that report fair/poor health in a community are the greatest indicators of that community's overall health. 

Next let us consider which parameters have the greatest correlation when using combined physically unhealthy and mentally unhealthy days, or the overall unhealthly days parameter as I will now refer to it,  as an indicator of community health.

```python
# Create new column to represent the percentage of days people reported unhealthy physically and/or mentally
df_pivot_new = df_pivot
df_pivot_new['% unhealthy days per month'] = (df_pivot['Physically Unhealthy Days']+df_pivot['Mentally Unhealthy Days'])/30


# Find correlation between each variable and the unhealthy days parameter.
corr_df2 = df_pivot_new.corr()
corr_df2 = corr_df2.drop('Physically Unhealthy Days')
corr_df2 = corr_df2.drop('Mentally Unhealthy Days')
unhealthy_days_corr_df = corr_df2[['% unhealthy days per month']]["% unhealthy days per month"]
unhealthy_days_corr_df = unhealthy_days_corr_df.drop('% unhealthy days per month')
unhealthy_days_corr_df = unhealthy_days_corr_df.sort_values()

# Choose the greatest correlation parameters from all parameters considered 
greatest_unhealthy_days_corr_df = unhealthy_days_corr_df[(unhealthy_days_corr_df >= .3) | (unhealthy_days_corr_df <= -.3)]
greatest_unhealthy_days_corr_df = greatest_unhealthy_days_corr_df.sort_values()

# Display the correlation values in order in a bar graph
# Plot all correlation values and the greatest correlation value parameters for % unhealthy days per month
fig, ax = plt.subplots(1,2, figsize=(15,5))

unhealthy_days_corr_df.plot.bar(title = 'Correlation Values for % of Unhealthy Days per Month',ax=ax[0])
greatest_unhealthy_days_corr_df.plot.bar(title = 'Greatest Value Correlation Parameters for % of Unhealthy Days per Month',ax=ax[1], color='red')
```

The two bar graphs above suggest that median family income, the percent of the population that is obese, the percent of the population that is unemployed, the poverty rate, the percentage of children in poverty, the percent of premature deaths in a population, the percentage of smokers and the percentage of individuals that report fair/poor health in a community are the greatest indicators of that community's overall health. 

Both health indicators, obesity rate and percentage of unhealthy days in a month, suggest that the same parameters have the greatest impact on a community's overall health.
